# **LED Lifetime Prediction V2**


## **0.Import Data**

```python
import pandas as pd
import numpy as np
```

```python
dataset=[]

import os
for root,dirs,files in os.walk("/content/"):
  for file in files:
       if file.endswith(".csv"):
          print(file)
          try:
            new=pd.read_csv(file,header=None,index_col=False,names=['wavelenght','A','B','C','D','E'])
            print(new)
            #split file name
            file=file[:-4].split("_")
            #drop restart
            if file[-1]=='restart':
              file=file[0:-1]
            #get mA
            mA=int(file[3][:-2])
            #get hr
            if 'o' in file[4][:-2]:
              spl=file[4][:-2].split("o")
              hr=int(spl[0])+int(spl[1])/10
            else:
              hr=int(file[4][:-2])
            #get tem
            tem=int(file[2][:-1])
            #only need first 6 column (some csv have 10 column)
            new=new.iloc[:,:6]
            new['tem']=tem
            new['mA']=mA
            new['hr']=hr
            dataset.append(new)
            print(new)
          except:
            print('no file')
          
dataset=pd.concat(dataset)
```



---
## **1. Reshape: normalize+concat**

```python
df=dataset
df.fillna(0)
```

```python
df_A=df.loc[:,['wavelenght','tem','mA','hr','A']]
df_A['LED']='A'
df_A.columns=['wavelenght','tem','mA','hr','Li','LED']
df_B=df.loc[:,['wavelenght','tem','mA','hr','B']]
df_B['LED']='B'
df_B.columns=['wavelenght','tem','mA','hr','Li','LED']
df_C=df.loc[:,['wavelenght','tem','mA','hr','C']]
df_C['LED']='C'
df_C.columns=['wavelenght','tem','mA','hr','Li','LED']
df_D=df.loc[:,['wavelenght','tem','mA','hr','D']]
df_D['LED']='D'
df_D.columns=['wavelenght','tem','mA','hr','Li','LED']
df_E=df.loc[:,['wavelenght','tem','mA','hr','E']]
df_E['LED']='E'
df_E.columns=['wavelenght','tem','mA','hr','Li','LED']

df=pd.concat([df_A,df_B,df_C,df_D,df_E])
df
```

```python
df.to_csv('/content/LI_combined.csv')
```

## **2. Add Feature: Mean of A-E**


```python
df=dataset
df=df.drop(index=779).fillna(0)
df['mean']= df.loc[:,['A','B','C','D','E']].mean(axis=1)
df[df['mean']!=0]
```

```python
df_A=df.loc[:,['wavelenght','tem','mA','hr','A']]
df_A['LED']='A'
df_A.columns=['wavelenght','tem','mA','hr','Li','LED']
df_B=df.loc[:,['wavelenght','tem','mA','hr','B']]
df_B['LED']='B'
df_B.columns=['wavelenght','tem','mA','hr','Li','LED']
df_C=df.loc[:,['wavelenght','tem','mA','hr','C']]
df_C['LED']='C'
df_C.columns=['wavelenght','tem','mA','hr','Li','LED']
df_D=df.loc[:,['wavelenght','tem','mA','hr','D']]
df_D['LED']='D'
df_D.columns=['wavelenght','tem','mA','hr','Li','LED']
df_E=df.loc[:,['wavelenght','tem','mA','hr','E']]
df_E['LED']='E'
df_E.columns=['wavelenght','tem','mA','hr','Li','LED']
df_M=df.loc[:,['wavelenght','tem','mA','hr','mean']]
df_M['LED']='Mean'
df_M.columns=['wavelenght','tem','mA','hr','Li','LED']

df=pd.concat([df_A,df_B,df_C,df_D,df_E,df_M])
df
```
